# Exercício

## 1. Apresentação gráfica da classificação

Com base nas classificações realizadas na lista de exercícios anteriores, implemente um código que apresente um gráfico que seja possível entender a qual classe cada instância pertence e quais instâncias foram classificadas certas e erradas.

Para isso, a instância de cada classe deve ser representada por um símbolo específico e estar na cor verde, se classificada corretamente, ou na cor vermelha, se tiver sido classificada de maneira errada.

Essa imagem é ilustrativa e para entendimento que as classes são definidas pelo símbolo usado e a corretude da classificação é identificada pela cor.
